**Taylor series** is a representation of a function as infinite sum of the function's derivatives at a single point.

With Taylor serires, we can use polynomial to descibe any *smooth function*. Polynomials are much easier to compute.  
Especially when we want to compute the value of $pi$, $e$ and trigonometric functions using computer.

Here, let's try to demonstrate the first order Taylor approximation using Python.

### Step 1: Import Packages

`numpy`: for mathematic functions.  
`matplotlib.pyplot`: for graph ploting.

```python
import numpy as np
from matplotlib import pyplot as plt
```

### Step 2: Define Function

Any funciton is ok.  
Here we choose a trigonometric function since its periodicity: $f(x) = 2\sin(x) + 3\cos(x) \cdot x - 1$

```python
def myFunction(x):
    f = 2*np.sin(x) + 3*np.cos(x)*x - 1
    return f
```

Let's see what the function image looks likeï¼š

```python
x = np.arange(-8*np.pi, 8*np.pi, 0.1) # set variable x's range and step
func = myFunction(x) # get our function

plt.figure(1)
plt.plot(x, func, 'r', label="function")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

### Step 3: Define Taylor Series of our function

#### 3.1 First Order Taylor Approximation

**Taylor series**:
$f(a) + \frac{f'(a)}{1!}(x-a) + \frac{f''(a)}{2!}(x-a)^2 + \frac{f'''(a)}{3!}(x-a)^3 + \cdots$
 
Because we just want to compute the first-order approximation, this is what we get:
$f(a) + \frac{f'(a)}{1!}(x-a)$

#### 3.2 Compute first order derivative of our function

$f'(x) = 5\cos(x) - 3\sin(x) \cdot x$

```python
def firstOrderDerivative(x):
    df = 5*np.cos(x) - 3*np.sin(x)*x
    return df
```

#### 3.3 Get First Order Taylor Approximation Function

$f(a) + \frac{f'(a)}{1!}(x-a) = 2\sin(a) + 3\cos(a) \cdot a - 1 + (5\cos(x) - 3\sin(x) \cdot x) \cdot (x - a)$

```python
def firstOrderTaylor(x, a):
    f = myFunction(a) # f(a)
    df = firstOrderDerivative(x) # f'(x)
    return f + df*(x-a)
```

### Plot the Graph

Now let's choose some points to observe the approximation.

#### Point One:  a = 0 

We got approximate function: $2\sin(0) + 3\cos(0) \cdot 0 - 1 + (5\cos(x) - 3\sin(x) \cdot x) \cdot (x - 0)$

```python
x0 = np.arange(-2*np.pi, 2*np.pi, 0.1) # change for better observation
f0 = firstOrderTaylor(x0, 0) # approximate function
func0 = myFunction(x0)

plt.figure(2)
plt.plot(x0, func0, 'r', label="function") # plot the original function
plt.plot(x0, f0, 'b', label="approximation around 0") # plot the approximation
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

#### Point Two: a = 7

We got approximate function: $2\sin(7) + 3\cos(7) \cdot 7 - 1 + (5\cos(x) - 3\sin(x) \cdot x) \cdot (x - 7)$

```python
x7 = np.arange(1*np.pi, 4*np.pi, 0.1) # change for better observation
f7 = firstOrderTaylor(x7, 7) # approximate function
func7 = myFunction(x7)

plt.figure(3)
plt.plot(x7, func7, 'r', label="function") # plot the original function
plt.plot(x7, f7, 'b', label="approximation around 7") # plot the approximation
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

#### Point Three: a = -5.5

We got approximate function: $2\sin(-5.5) + 3\cos(-5.5) \cdot (-5.5) - 1 + (5\cos(x) - 3\sin(x) \cdot x) \cdot (x + 5.5)$

```python
x5 = np.arange(-3*np.pi, np.pi, 0.1) # change for better observation
f5 = firstOrderTaylor(x5, -5.5) # approximate function
func5 = myFunction(x5)

plt.figure(4)
plt.plot(x5, func5, 'r', label="function") # plot the original function
plt.plot(x5, f5, 'b', label="approximation around -5.5") # plot the approximation
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

We can see from previous figures that original function and its approximate function are very close near the given point.  
But after leaving a certain area, the correlation between the two disappeared.

### Second Order Taylor Approximation

If we want to get a more precise approximation, we can add higher order derivatives.  
Let's add the second order derrivative of our function to see the result.

Everything we did before are still same. The only thing we have to do is add the second order derivative of the function.

#### Second Order Derivative

We have already got the first order derivative: $f'(x) = 5\cos(x) - 3\sin(x) \cdot x$  
Differentiate the function again, we get: $f''(x) = -8\sin(x) - 3\cos(x) \cdot x$

#### Second Order Taylor Approximation Function

$f(a) + \frac{f'(a)}{1!}(x-a) + \frac{f''(a)}{2!}(x-a)^2
= 2\sin(a) + 3\cos(a) \cdot a - 1
+ (5\cos(x) - 3\sin(x) \cdot x) \cdot (x - a)
+ \frac{-8\sin(x) - 3\cos(x) \cdot x}{2} \cdot (x-a)^2$

```python
def secondOrderDerivative(x):
    f = -8*np.sin(x) - 3*np.cos(x)*x
    return f

def secondOrderTaylor(x, a):
    firstOrder = firstOrderTaylor(x, a)
    sd = secondOrderDerivative(x)
    secondOrder = firstOrder +  sd/2 * (x-a)**2
    
    return secondOrder
```

Now let's plot the graph of `a = 0` again:

```python
x_s0 = np.arange(-1*np.pi, 1*np.pi, 0.1)
second_0 = secondOrderTaylor(x_s0, 0)
func_s0 = myFunction(x_s0)

plt.figure(5)
plt.plot(x_s0, func_s0, 'r', label="function") # plot the original function
plt.plot(x_s0, second_0, 'b', label="approximation around 0") # plot the approximation
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
